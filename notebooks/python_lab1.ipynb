```python
import numpy as np
import typing
import pandas as pd
file = pd.read_csv("/content/olist_order_items_dataset.csv")
'''Выделяем столбцы с ценой и доставкой'''
py_list = file[['price', 'freight_value']]
```

**1 задание**

```python
def realisation_a(py_list: pd.Series) -> list:
  '''Простой Python
  Принимает на вход таблицу py_list
  Возвращает финальную цену, полученную по формуле, - final_price
  '''
  final_price = []

  '''Проходим по столбцам и выполняем необходимые операции'''
  for i, row in py_list.iterrows():
    price = row["price"]
    freight_value = row["freight_value"]
    discount = 0
    if price > 500:
      discount = 0.15
    elif price > 100:
      discount = 0.1
    fin_curr_price = (price * (1 - discount)) + freight_value
    final_price.append(fin_curr_price)
  return final_price
```

```python
'''Numpy'''
def realisation_b(price: np.ndarray, freight_value: np.ndarray) -> np.ndarray:
  '''Выделяем условия, выборы и значения скидок'''
  conditions = [price > 500, price > 100]
  big_disc = 0.15
  small_disc = 0.1
  choises = [(price * (1 - big_disc)) + freight_value,
             (price * (1 - small_disc)) + freight_value]
  '''Создаём массив с финальными ценами'''
  fin_price = np.select(conditions, choises, default=price + freight_value)
  return fin_price
```

```python
price = np.array(file['price'])
freight_value = np.array(file['freight_value'])
%timeit realisation_a(py_list)
%timeit realisation_b(price, freight_value)

'''
Время выполнения программы способа A - 4.79 секунд +- 580 милисекунд
Время выполнения программы способа B - 2.32 милисекунды += 84 микросекунды
'''
%who
difference = 3.63 * 1000 / 1.87
print(difference)
'''Время выполнения программы увеличилось в 1941 раза'''
```

```python
'''Получаю ненормированную матрицу с весом, высотой и тд'''
matr_not_normed = np.array(products[['product_length_cm', 'product_height_cm', 'product_width_cm', 'product_weight_g']])[:1000]
'''
Расчитываю min и max для формулы нормализации:
x_norm = (x - min(x)) / (max(x) - min(x))
'''
mins = np.array([min(matr_not_normed[:, 0]), min(matr_not_normed[:, 1]), min(matr_not_normed[:, 2]), min(matr_not_normed[:, 3])])
maxs = np.array([max(matr_not_normed[:, 0]), max(matr_not_normed[:, 1]), max(matr_not_normed[:, 2]), max(matr_not_normed[:, 3])])
matrix = np.multiply(matr_not_normed - mins, 1/(maxs-mins))
'''Для каждого вектора матрицы считаю норму и выделяю рандомный вектор'''
norms = np.linalg.norm(matrix, axis=1)
random_id = np.random.randint(matrix.shape[0])
random_product = matrix[random_id]
random_product_norm = np.linalg.norm(random_product)
'''Расчитываю косинусное сходноство по формуле a * b / (||a|| * ||b||)'''
cos_sim_no_norm = np.dot(matrix, random_product)
cos_sim = np.multiply(cos_sim_no_norm, 1 / norms) / random_product_norm
print(random_id, matr_not_normed[random_id])
print(cos_sim)
```
