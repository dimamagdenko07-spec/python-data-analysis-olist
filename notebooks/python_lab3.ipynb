```python
import pandas as pd
import numpy as np
orders = pd.read_csv('olist_orders_dataset.csv')
customers = pd.read_csv('olist_customers_dataset.csv')
items = pd.read_csv('olist_order_items_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
translation = pd.read_csv('product_category_name_translation.csv')
```

Задание 1

```python
'''Сливаем таблицы'''
df = orders.merge(items, on='order_id', how='left')
'''on='order_id' значит, что во всем таблицах поиск идёт по order_id
это - связующий элемент'''
'''how=left значит, что мы оставляем все строки в левой таблице,
даже если для них нет соответствий в правой'''
df = df.merge(customers, on='customer_id', how='left')
df = df.merge(products, on='product_id', how='left')
df = df.merge(translation, on='product_category_name', how='left')
df['product_category_name_english'] = df['product_category_name_english'].fillna('unknown')


'''Оставляем доставленные и удаляем места, где не можем считать цену'''
delivered = df[df['order_status'] == 'delivered'].copy()
delivered = delivered.dropna(subset=['price'])

'''Расчитываем стоимость'''
delivered['total_cost_of_item'] = delivered['price'] + delivered['freight_value']
total_order_cost = delivered.groupby('order_id')['total_cost_of_item'].sum().reset_index()
total_order_cost.rename(columns={'total_cost_of_item' : 'total_order_cost'}, inplace=True)
'''группируем товары из одного заказа и суммируем для них суммарную цену'''
delivered = delivered.merge(total_order_cost, on='order_id', how='left')
df.to_csv('merged.csv')

display(df)
```

Задание 2

```python
delivered['order_purchase_timestamp'] = pd.to_datetime(delivered['order_purchase_timestamp'])
today = (delivered['order_purchase_timestamp'].max()) + pd.Timedelta(days=1)
rfm = delivered.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (today - x.max()).days,
    'order_id' : 'nunique',
    'total_cost_of_item' : 'sum'
}).reset_index()
rfm.columns = ['customer_unique_id', 'Recency', 'Frequency', 'Monetary']
states = delivered[['customer_unique_id', 'customer_state']].drop_duplicates()
rfm = rfm.merge(states, on='customer_unique_id', how='left')
rfm['R_status'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1])
rfm['F_status'] = pd.cut(rfm['Frequency'], [0,1,2,3, 1000], labels=[1,2,3,4])
rfm['M_status'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4])
rfm['Status'] = rfm['R_status'].astype(str) + rfm['F_status'].astype(str) + rfm['M_status'].astype(str)
print(rfm[rfm['Status'] == '444']['customer_state'].value_counts().head(5))
```

Задание 3

```python
top_categories = delivered.groupby('product_category_name_english')['total_cost_of_item'].sum().nlargest(10).index
df_top = delivered[delivered['product_category_name_english'].isin(top_categories)]
matrix = df_top.pivot_table(
    index='customer_state',
    columns='product_category_name_english',
    values='total_cost_of_item',
    aggfunc='sum'
).fillna(0)

matrix = matrix.div(matrix.sum(axis=1), axis=0)
mean_hb = matrix['health_beauty'].mean()
std_nb = np.std(matrix['health_beauty'])
z = (matrix['health_beauty'] - mean_hb)/std_nb
print(matrix[abs(z) > 1.8])
```

Задание 4

```python
delivered['month-year'] = delivered['order_purchase_timestamp'].dt.to_period('M')
delivered['cohort_month'] = delivered.groupby('customer_unique_id')['month-year'].transform('min')
#print(delivered['cohort_month'])

year = delivered['month-year'].dt.year - delivered['cohort_month'].dt.year
month = delivered['month-year'].dt.month - delivered['cohort_month'].dt.month
delivered['month_index'] = year * 12 + month
cohort = delivered.groupby(['cohort_month', 'month_index'])['customer_unique_id'].nunique().reset_index()
cohort_table = cohort.pivot_table(
    index='cohort_month',
    columns='month_index',
    values = 'customer_unique_id'
).fillna(0)
full = cohort_table[0]
cohort_table = cohort_table.div(full, axis=0)
cohort_table.to_csv('cohort_analysis.csv')
print(cohort_table)
```
