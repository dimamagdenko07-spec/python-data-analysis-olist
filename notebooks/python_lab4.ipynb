Задание 1

```python
import pandas as pd
import numpy as np

orders = pd.read_csv('olist_orders_dataset.csv')
time = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
'''Преобразуем даты'''
for col in time:
  orders[col] = pd.to_datetime(orders[col])
display(orders)
indexed_orders = orders.copy()
indexed_orders.set_index('order_purchase_timestamp', inplace = True)
```

```python
'''Считаем, сколько дней прошло с момента покупки до момента доставки'''
df = indexed_orders.dropna(subset=['order_approved_at', 'order_delivered_customer_date']).copy()
start_days = df['order_approved_at'].values.astype('datetime64[D]')
end_days = df['order_delivered_customer_date'].values.astype('datetime64[D]')

df['time_days'] = np.busday_count(start_days, end_days)
df.to_csv('table_with_days.csv')
display(df)
```

```python
'''Смотрим на просроченные товары'''
indexed_orders['is_delayed'] = indexed_orders['order_delivered_customer_date'] > indexed_orders['order_estimated_delivery_date']
'''Смотрим процент просроченных товаров в месяце'''
delay_mean = indexed_orders['is_delayed'].resample('MS').mean() * 100
max_delay_month = delay_mean.idxmax()
max_delay_percent = delay_mean.max()
display(indexed_orders)
indexed_orders.to_csv("prod.csv")
print("Месяц с максимальным процентов просроченных заказов: ", max_delay_month)
print("Процент просроченных заказов в этот месяц: ", max_delay_percent)
```

Задание 2

```python
items = pd.read_csv('olist_order_items_dataset.csv')
'''Сливаем цену и время покупки по индексу заказа '''
df_price = pd.merge(orders[['order_id', 'order_purchase_timestamp']], items[['order_id', 'price']], on='order_id')
'''Устанавливаем время покупки индексом и группируем по датам(создавая пропущенные дни)'''
days = df_price.set_index('order_purchase_timestamp')['price'].resample('D').sum()
days.to_csv("days.csv")
display(days)
```

```python
df_price['weekday'] = df_price['order_purchase_timestamp'].dt.day_name()
df_price['hours'] = df_price['order_purchase_timestamp'].dt.hour
'''Смотрим в какое время у нас продаётся больше товаров каждый день'''
table = df_price.pivot_table(
    index='weekday',
    columns='hours',
    values='order_id',
    aggfunc='count'
    )
table.to_csv('intraday_pattern.csv')
peaks = table.idxmax(axis=1)
days_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
peaks = peaks.reindex(days_week)
display(peaks)
```

Задание 3

```python
'''Считаем кумулятивную выручки с самого начала по сегодняшний день'''
days_rolling = days.groupby(days.index.year).expanding().sum()

days_rolling = days_rolling.reset_index(level=0, drop=True)
'''Считаем скользящее среднее и скользящее стандартное отклонение за 30 дней и ищем аномалии'''
rolling_mean = days.rolling(30).mean()
rolling_std = days.rolling(30).std()
Z = (days - rolling_mean) / rolling_std
anomalies = Z[Z>3]
Z.to_csv('z_score.csv')
display(Z)
```

Задание 4

```python
'''Доход за последние 7 дней, не считая этот'''
revenue_last_7d = days.shift(1).rolling(7).sum()
'''Доход за неделю до этой недели'''
revenue_prev_7d = days.shift(7).rolling(7).sum()
'''Считаем относительный темп'''
wow_growth = (revenue_last_7d - revenue_prev_7d) / revenue_prev_7d * 100

print(wow_growth)
'''Ищем набольшие падения в доходе'''
biggest_drops = wow_growth.nsmallest(3)
print(biggest_drops)
```
